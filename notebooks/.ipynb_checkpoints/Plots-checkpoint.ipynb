```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MaxNLocator
```

```python
df = pd.read_csv('../patient_features.csv')
cgm_df = pd.read_csv('../test.csv')
# cgm_df
# len(df[(df.patient_number == 3)].values)
df
```

```python
df
```

```python
# plots for longest strike above mean for each patient
for i in range(1,6):
    X = df[df.patient_number == i]['meal_number'].values
    Y = df[df.patient_number == i]['lsam'].values
    ax = plt.figure().gca()
    ax.xaxis.set_major_locator(MaxNLocator(integer = True))
    plt.scatter(X, Y, label='lsam')
    
    plt.plot(X, Y)
    plt.xlabel('Meal Number')
    plt.ylabel('Feature Value')
    plt.legend()
    plt.title("lsam feature for patient " + str(i))
    # plt.show()
    plt.savefig("plots/lsam/pat_" + str(i) + "_lsam" + ".png", dpi=200)
    plt.close()
```

```python
# plots for max min_max difference for each patient
for i in range(1,6):
    X = df[df.patient_number == i]['meal_number'].values
    Y = df[df.patient_number == i]['min_max'].values
    ax = plt.figure().gca()
    ax.xaxis.set_major_locator(MaxNLocator(integer = True))
    plt.scatter(X, Y, label='min_max')
    
    plt.plot(X, Y)
    plt.xlabel('Meal Number')
    plt.ylabel('Feature Value')
    plt.legend()
    plt.title("min_max feature for patient " + str(i))
    # plt.show()
    plt.savefig("plots/min_max/pat_" + str(i) + "_min_max" + ".png", dpi=200)
    plt.close()
```
